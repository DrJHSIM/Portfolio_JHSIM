```python
# 라이브러리와 데이터를 불러오고, 시각화를 위한 세팅을 합니다.
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("penguins")
data = data.dropna()
```

```python
data
```

### Histplot 


```python
# penguin 데이터에 histplot을 출력합니다.
plt.figure(figsize=(6, 6))
plt.title("Distribution over bill_length_mm", fontsize=14)
sns.histplot(data=data, x="bill_length_mm", bins=15, hue="species", multiple="stack")
plt.show()
```

### Displot 


```python
# penguin 데이터에 displot을 출력합니다.
sns.displot(data=data, kind="kde", x="bill_length_mm", hue="species", col="island")
plt.show()
```

### Barplot


```python
# penguin 데이터에 barplot을 출력합니다.
#sns.barplot(data=data, x="species", y="body_mass_g")
sns.barplot(data=data, x="bill_length_mm", y="species", hue="sex")
plt.show()
```

### Countplot


```python
# penguin 데이터에 countplot을 출력합니다.
plt.figure(figsize=(8, 6))
sns.set_palette("viridis")
sns.countplot(data=data, x="sex", hue="species")
plt.show()
```

### Boxplot



```python
# penguin 데이터에 boxplot을 출력합니다.
plt.figure(figsize=(8, 6))
sns.set_palette("Set2")
sns.boxplot(data=data, y="species", x="bill_length_mm", hue="sex")
plt.show()
# What is boxplot?
```

### Violinplot 


```python
# penguin 데이터에 violinplot을 출력합니다.
plt.figure(figsize=(8, 6))
sns.violinplot(data=data, y="species", x="bill_length_mm", hue="sex")
plt.show()
```

### Lineplot


```python
# penguin 데이터에 lineplot을 출력합니다.
plt.figure(figsize=(8, 6))
sns.lineplot(data=data, x="body_mass_g", y="bill_length_mm", ci=None)
plt.show()
```

### Pointplot


```python
# penguin 데이터에 pointplot을 출력합니다.
plt.figure(figsize=(8, 6))
sns.pointplot(data=data, x="island", y="bill_length_mm")
plt.show()
```

### Scatterplot



```python
# penguin 데이터에 scatterplot을 출력합니다.
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x="flipper_length_mm", y="bill_length_mm", hue="sex")
plt.show()
```

### Pairplot


```python
# penguin 데이터에 pairplot을 출력합니다.
plt.figure(figsize=(10, 10))
sns.pairplot(data=data, hue="sex")
plt.show()
```

### Heatmap


**상관관계란?**
- 어떤 X값의 변화에 따라 Y값의 선형적으로 변화하는지를 측정한 지표.

- [-1, 1]


e.g. 아이스크림 판매량 증가 <----> 상어에 물린 사람 수

```python
# 각 feature간 상관관계를 파악하기 위해 Correlation matrix를 만듭니다.
corr = data.corr()
corr
```

```python
# penguin 데이터에 heatmap을 출력합니다.
plt.figure(figsize=(8, 8))
sns.heatmap(data=corr, square=True, cmap="Blues", annot=True, fmt=".3f")
plt.show()
```
